### Αξιολόγηση Αποτελεσμάτων

Μετά την αποκατάσταση κάθε εικόνας, μπορείτε να συγκρίνετε την αποκατεστημένη εικόνα με την αρχική `lena.png`, τόσο οπτικά/ποιοτικά, όσο και ποσοτικά, χρησιμοποιώντας δείκτες αξιολόγησης όπως MSE, PSNR και SSIM. Ένα ενδεικτικό τμήμα κώδικα (σε σχόλια) παρέχεται σε κάθε ερώτημα για να υπολογίσετε αυτούς τους δείκτες.

- **MSE (Mean Squared Error)**: Μετρά το μέσο τετράγωνο της διαφοράς pixel ανάμεσα στις εικόνες. Όσο μικρότερη η τιμή, τόσο καλύτερη η αποκατάσταση.
- **PSNR (Peak Signal-to-Noise Ratio)**: Εκφράζει την ποιότητα της εικόνας σε dB. Όσο μεγαλύτερη η τιμή της σηματοθορυβικής σχέσης, τόσο καλύτερη η αποκατάσταση.
- **SSIM (Structural Similarity Index Measure)**: Μετρά τη δομική ομοιότητα των εικόνων, λαμβάνοντας υπόψη την αντίθεση, φωτεινότητα και υφή. Τιμές κοντά στο 1 δείχνουν υψηλή ομοιότητα.

### Helper για Histogram Specification (Καθορισμός Ιστογράμματος)

Στο πλαίσιο της άσκησης, θα χρειαστεί να εντοπίσετε ποια από τις επτά παραμορφωμένες εικόνες έχει υποστεί **ανεπαρκή ισοστάθμιση ιστογράμματος (poor histogram equalization)**. Για την περίπτωση αυτή, σας παρέχεται το παρακάτω βοηθητικό μπλοκ κώδικα που υπολογίζει την τεχνική του **καθορισμού ιστογράμματος (histogram specification)**, χρησιμοποιώντας την εικόνα-στόχο `lena.png` ως πηγή για το επιθυμητό ιστόγραμμα.

Θα χρειαστεί να **προσαρμόσετε κατάλληλα** τον κώδικα εντός του σωστού κελιού (ανάλογα με το distortion), ώστε να επιτύχετε την αποκατάσταση.

#### Περιγραφή Μεθόδου

Η διαδικασία περιλαμβάνει:

1. **Υπολογισμό αθροιστικής κατανομής (CDF)** της παραμορφωμένης εικόνας και του reference ιστογράμματος.
2. **Εντοπισμό mapping** τιμών μέσω αντιστοίχισης των CDF: για κάθε τιμή φωτεινότητας στην παραμορφωμένη εικόνα, αντιστοιχίζεται η πλησιέστερη τιμή του reference CDF (βασισμένο στην εξίσωση CDF matching).
3. **Εφαρμογή του mapping** στην εικόνα για να δημιουργηθεί η νέα εικόνα με την επιθυμητή κατανομή φωτεινότητας.

**Σημείωση:** Η υλοποίηση βασίζεται σε `np.interp()` για κατασκευή του mapping πίνακα, ακολουθώντας την αρχή του αντιστρόφου μετασχηματισμού CDF. Παρά το γεγονός ότι δεν παράγεται τέλεια προσαρμογή, το αποτέλεσμα είναι κατά κανόνα σημαντικά βελτιωμένο.

---

```python
# Το παρακάτω block σας βοηθά να αντιμετωπίσετε την περίπτωση παραμόρφωσης με ανεπαρκή ισοστάθμιση ιστογράμματος.
# Μην το εκτελέσετε όπως είναι — χρησιμοποιήστε το ως οδηγό και προσαρμόστε το σωστά στο κελί που αντιστοιχεί
# στο κατάλληλο distortion.

# === 1. Υπολογισμός reference ιστογράμματος από την αρχική εικόνα ===
# Χρησιμοποιείται ως "στόχος" για την ανακατανομή των εντάσεων
hist_ref, _ = np.histogram(target.flatten(), bins=256, range=(0, 256))
hist_ref = hist_ref.astype(np.float64) / hist_ref.sum()  # Κανονικοποίηση ώστε να αναπαριστά πιθανότητες

# === 2. Συνάρτηση: Καθορισμός Ιστογράμματος μέσω CDF Matching ===
def histogram_specification(source_img, reference_hist):
    # Βήμα 1: Υπολογισμός ιστογράμματος της εισόδου
    src_hist, _ = np.histogram(source_img.flatten(), bins=256, range=(0, 256))
    src_cdf = np.cumsum(src_hist).astype(np.float64)
    src_cdf /= src_cdf[-1]  # Κανονικοποίηση στο [0, 1]

    # Βήμα 2: Υπολογισμός CDF της reference κατανομής
    ref_cdf = np.cumsum(reference_hist).astype(np.float64)
    ref_cdf /= ref_cdf[-1]

    # Βήμα 3: Δημιουργία mapping s_k -> z_q μέσω πλησιέστερης τιμής CDF
    mapping = np.interp(src_cdf, ref_cdf, np.arange(256))

    # Βήμα 4: Εφαρμογή mapping στα pixels της εικόνας
    matched = np.interp(source_img.flatten(), np.arange(256), mapping)
    return matched.reshape(source_img.shape).astype(np.uint8)

# === 3. Παράδειγμα χρήσης (συμπληρώστε με τον σωστό αριθμό εικόνας) ===
# Η εικόνα με την ανεπαρκή ισοστάθμιση ιστογράμματος είναι η 1 καθώς είναι αρκετά φωτεινή, δηλαδή οι τιμές του ιστογράμματος
# είναι συγκεντρωμένες σε ένα στενό εύρος δεξιά και άρα υπα΄ρχει ανεπαρκής ισοστάθμιση ιστογράμματος.
    
    input_img = distorted_images["Distorted Image #1"].copy()
# restored = histogram_specification(input_img, hist_ref)

# === 4. Οπτική αξιολόγηση (προαιρετικά) ===
    show_image("Distorted Image #1", input_img)
    show_image("Restored Image", restored)

# === 5. Συγκριτική απεικόνιση ιστογραμμάτων (προαιρετικά) ===
    hist_input, _ = np.histogram(input_img.flatten(), bins=256, range=(0, 256))
    hist_input = hist_input.astype(np.float64) / hist_input.sum()
    hist_restored, _ = np.histogram(restored.flatten(), bins=256, range=(0, 256))
    hist_restored = hist_restored.astype(np.float64) / hist_restored.sum()
    hist_target, _ = np.histogram(target.flatten(), bins=256, range=(0, 256))
    hist_target = hist_target.astype(np.float64) / hist_target.sum()
    plt.figure(figsize=(10, 4))
    plt.plot(hist_input, label='Distorted Image', color='gray')
    plt.plot(hist_restored, label='Restored Image', color='green', linestyle='--')
    plt.plot(hist_target, label='Original Lena', color='red', linestyle=':')
    plt.title("Overlay of Histograms")
    plt.xlabel("Intensity")
    plt.ylabel("Normalized Frequency")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
## Αποκατάσταση: Distorted Image #1
# Η εικόνα με την ανεπαρκή ισοστάθμιση ιστογράμματος είναι η 1 καθώς είναι αρκετά φωτεινή, δηλαδή οι τιμές του ιστογράμματος
# είναι συγκεντρωμένες σε ένα στενό εύρος δεξιά και άρα υπα΄ρχει ανεπαρκής ισοστάθμιση ιστογράμματος.

input_img = distorted_images["Distorted Image #1"].copy()

# === Πριν την αποκατάσταση: Αξιολόγηση "distorted" ===
mse_distorted = np.mean((target - input_img) ** 2)
psnr_distorted = psnr(target, input_img)
ssim_distorted = ssim(target, input_img)
print("Before Restoration:")
print("MSE:", mse_distorted)
print("PSNR:", psnr_distorted)
print("SSIM:", ssim_distorted)

# === Εφαρμογή τεχνικών αποκατάστασης ===
hist_ref, _ = np.histogram(target.flatten(), bins=256, range=(0, 256))
hist_ref = hist_ref.astype(np.float64) / hist_ref.sum()  

# === 2. Συνάρτηση: Καθορισμός Ιστογράμματος μέσω CDF Matching ===
def histogram_specification(source_img, reference_hist):
    src_hist, _ = np.histogram(source_img.flatten(), bins=256, range=(0, 256))
    src_cdf = np.cumsum(src_hist).astype(np.float64)
    src_cdf /= src_cdf[-1] 

    # Βήμα 2: Υπολογισμός CDF της reference κατανομής
    ref_cdf = np.cumsum(reference_hist).astype(np.float64)
    ref_cdf /= ref_cdf[-1]

    # Βήμα 3: Δημιουργία mapping s_k -> z_q μέσω πλησιέστερης τιμής CDF
    mapping = np.interp(src_cdf, ref_cdf, np.arange(256))

    # Βήμα 4: Εφαρμογή mapping στα pixels της εικόνας
    matched = np.interp(source_img.flatten(), np.arange(256), mapping)
    return matched.reshape(source_img.shape).astype(np.uint8)

restored = histogram_specification(input_img, hist_ref)
# === Οπτική αξιολόγηση ===
show_image("Distorted Image #1", input_img)
show_image("Restored Image", restored)

# === Ποσοτική αξιολόγηση μετά την αποκατάσταση (με χρήση της lena.jpg ως ground truth) ===
mse_score = np.mean((target - restored) ** 2)
psnr_score = psnr(target, restored)
ssim_score = ssim(target, restored)
print("After Restoration:")
print("MSE:", mse_score)
print("PSNR:", psnr_score)
print("SSIM:", ssim_score)
```

```python
## Αποκατάσταση: Distorted Image #2
# Η εικόνα 2 έχει υποστεί θόλωση με γραμμικό φιλτράρισμα με φίλτρο μέσου 5x5 (Mean Blur), καθώς φαίνεται πως η θολωσή της είναι μεγαλύτερη από
# την εικόνα 3, και θα εφαρμόσω την Όξυνση (Sharpening) για την αποκαταστασ΄η της. Ο λόγος για τον οποίο μόνο ο δείκτης MSE παρουσιάζει σημαντική 
# βελτίωση είναι δίοτι η όξυνση επηρεάζει κυρίως τις ακμές, επαναφέροντας αριθμητικά τις τιμές των pixel πιο κοντά στην αρχική εικόνα, γι'αυτό μειώνεται.
from scipy.signal import convolve2d
input_img = distorted_images["Distorted Image #2"].copy()

# === Πριν την αποκατάσταση: Αξιολόγηση "distorted" ===
mse_distorted = np.mean((target - input_img) ** 2)
psnr_distorted = psnr(target, input_img)
ssim_distorted = ssim(target, input_img)
print("Before Restoration:")
print("MSE:", mse_distorted)
print("PSNR:", psnr_distorted)
print("SSIM:", ssim_distorted)

# === Εφαρμογή τεχνικών αποκατάστασης ===
kernel = np.array([   # Πυρήνας Λαπλασιανής
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
def sharpening(source_img, kernel):
    sharp = convolve2d(source_img, kernel, mode = 'same', boundary = 'symm') # Εφαρμόζει 2D συνέλιξη της εικόνας με το πυρήνα, mode = 'same' σημαίνει
        # πως η έξοδος έχει ίδιο μέγεθος με την είσοδο, boundary = 'sym' χρησιμοποιείται για συμμετρική επέκταση των άκρων της εικόνας
    sharp = np.clip(sharp, 0, 255).astype(np.uint8) # Κανονικοποίηση και μετατροπή του τύπου
    return sharp  # Επιστροφή της οξυμένης εικόνας
restored = sharpening(input_img, kernel)  # Θέτει restored την οξυμένη εικόνα

# === Οπτική αξιολόγηση ===
show_image("Distorted Image #2", input_img)
show_image("Restored Image", restored)

# === Ποσοτική αξιολόγηση μετά την αποκατάσταση (με χρήση της lena.jpg ως ground truth) ===
mse_score = np.mean((target - restored) ** 2)
psnr_score = psnr(target, restored)
ssim_score = ssim(target, restored)
print("After Restoration:")
print("MSE:", mse_score)
print("PSNR:", psnr_score)
print("SSIM:", ssim_score)
```

```python
## Αποκατάσταση: Distorted Image #3
# Η εικόνα 3 έχει υποστεί εξομάλυνση με Γκαουσιανό φίλτρο 9x9 (Gaussian Blur) καi θα εφαρμόσω αφαίρεση εξομάλυνσης (Unsharp Masking) για
# αποκατάσταση της εικόνας. Παρατηρώ όμως πως με την αφαίρεση της θολόυρας με τη χρήση της μάσκας δεν αποκαθιστά πλήρως την εικόνα και παρατηρώ πως
# η εικόνα έχει θόρυβο αλατοπίπερου όπως η εικόνα 7 , έτσι πήρα το κώδικα απο την 7 ώστε η παραμορφωμένη εικόνα που προκύπτει να μπει σαν όρισμα
# στη συνάρτηση median_filtering για απομάκρυνση του θορύβου του αλατοπίπερου κι έτσι τελικά προκύπτει μια πολύ βελτιωμένη εικόνα.
input_img = distorted_images["Distorted Image #3"].copy()

# === Πριν την αποκατάσταση: Αξιολόγηση "distorted" ===
mse_distorted = np.mean((target - input_img) ** 2)
psnr_distorted = psnr(target, input_img)
ssim_distorted = ssim(target, input_img)
print("Before Restoration:")
print("MSE:", mse_distorted)
print("PSNR:", psnr_distorted)
print("SSIM:", ssim_distorted)

# === Εφαρμογή τεχνικών αποκατάστασης ===
ksize = (5, 5) # Πυρήνας Gaussian θολ΄ώματος 
sigma = 0.5    # Συντελεστής ενίσχυσης μάσκας
def unsharp_masking(source_img, ksize, sigma):
    blur = cv2.GaussianBlur(source_img, ksize, sigma)  # Θόλωση πυρήνα 
    mask = img.astype(np.float32) - blur.astype(np.float32)  # Υπολογισμός μάσκας, δηλαδή διαφορές αρχικής εικόνας με τη θολωμένη εικόνα 
    sharp = img.astype(np.float32) + sigma * mask  # Βάζει τη μάσκα στην αρχική εικόνα για ενίσχυση ακμών 
    sharp = np.clip(sharp, 0, 255).astype(np.uint8)  # Μετατροπή εικόνας ξανά σε τιμές 0-255 με uint8 
    return sharp
restored1 = unsharp_masking(input_img, ksize, sigma)  # Θέτω restored1 την προσωρινή τροποποιημένη εικόνας

window_size = 3   # Παράθυρο που είναι υποπεριοχή της εικόνας γύρω από κα΄θε pixel επεξεργασίας
def median_filtering(restored1, window_size): # Παίρνει σαν όρισμα την ήδη τροποποιημένη εικόνα από πριν
    pad = window_size // 2   # Κάνω ακεραία διαίρεση γιατί το padding πρέπει να είναι ακέραιο
    padded = np.pad(restored1, pad, mode='edge')  # Αναπαραγωγή άκρων της εικόνας
    filtered = np.zeros_like(restored1) # Δημιουργώ νέο πίνακα (με ίδιες διαστάσεις) με μηδενικές τιμές ώστε να αποθηκεύω σε αυτόν την νέα εικόνα με 
                            # το φίλτρο median και έπειτα διατρέχω κάθε pixel της εικόνας και το αποθηκεύω στη νέα εικόνα
    for i in range(restored1.shape[0]):
        for j in range(restored1.shape[1]):
            window = padded[i:i + window_size, j:j + window_size] # Παράθυρο για το τρέχον pixel και τα γειτονικά του
            median_value = np.median(window)  # Υπολογισμός διαμέσου των τιμών του pixel στο παράθυρο ( αγνοεί ακραίες τιμές, 0 ή 255, δηλαδή οι τιμές
                                            # για το μαύρο και το άσπρο του θόρυβου του αλατοπίπερου)
            filtered[i, j] = median_value  # Αντικατάσταση τιμών i, j του pixel με τη διάμεσο του παραθύρου
    return filtered  # Επιστρέφω την εικόνα μετά την αποκατάσταση αφού διέτρεξε για όλα τα pixel
restored2 = median_filtering(restored1, window_size) # Θέτω το restored με τη νέα εικόνα ώστε να τυπωθεί

# === Οπτική αξιολόγηση ===
show_image("Distorted Image #3", input_img)
show_image("Restored Image", restored2) # Αλλάζω τα restored από πριν με restored2 πού είναι η τελική μου εικόνα αποκατάστασης

# === Ποσοτική αξιολόγηση μετά την αποκατάσταση (με χρήση της lena.jpg ως ground truth) ===
mse_score = np.mean((target - restored2) ** 2)
psnr_score = psnr(target, restored2)
ssim_score = ssim(target, restored2)
print("After Restoration:")
print("MSE:", mse_score)
print("PSNR:", psnr_score)
print("SSIM:", ssim_score)
```

```python
## Αποκατάσταση: Distorted Image #4
#Η εικόνα έχει επεκταμένη αντίθεση(Contrast Stretching / Overstretching), δηλαδή οι φωτεινότητες έχουν απλωθεί αρκε΄τα στο εύρος 0-255 του ιστογράμματος
# με αποτέλεσμα να είναι υπερβολικά φωτεινή και υπερβολικά σκοτεινή σε κάποια σημεία. Για την αποκατάσταση της θα χρησιμοποιήσω περιορισμό εύρους
# τιμών μέσω σημειακών τελεστών (Log / Power Transformations) με στόχο να επαναφέρει τις τιμές στα μεσαία του εύρους. Επίσης, θα δημιουργήσω το
# ιστόγραμμα της Distorted Image για διευκόλυνση.
input_img = distorted_images["Distorted Image #4"].copy()

# === Εξαγωγή Ιστογράμματος ===
plt.hist(input_img.ravel(), bins = 256, range = (0, 255), color = "gray")  # Δημιουργία Ιστογράμματος με εύρος 0-255
plt.title("Histogram of Distorted Image #4")  # Τίτλος ιστογράμματος
plt.xlabel("Pixel Intensity")  # Ονομασία άξονα x
plt.ylabel("Frequency")        # Ονομασία άξονα y
plt.grid(True)  # Δημιουργία πλέγματος 
plt.show   # Απεικόνιση ιστογράμματος

# === Πριν την αποκατάσταση: Αξιολόγηση "distorted" ===
mse_distorted = np.mean((target - input_img) ** 2)
psnr_distorted = psnr(target, input_img)
ssim_distorted = ssim(target, input_img)
print("Before Restoration:")
print("MSE:", mse_distorted)
print("PSNR:", psnr_distorted)
print("SSIM:", ssim_distorted)

# === Εφαρμογή τεχνικών αποκατάστασης ===
x = 0.85  # Παράμετρος που δείχνει πόσο συμπιέζω ή απλώνω τις τιμές στο ιστόγραμμα, ισχύει και για τις 2 λύσεις
def power_transformation(source_img, x):
    normalized = source_img / 255  # Μετατροπή εικόνας απο εύρος 0-255 σε 0-1 για ευκολία στην ύψωση στη δύναμη στο gamma παρακάτω
    #transform = np.power(normalized, x) # Λύση 1, κάθε τιμή φωτεινότητας υψώνεται στο gamma. Παρατηρώ πως με αυτή τη μέθοδο δυσκολεύομαι
    # να βρω καλό αποτέλεσμα και το αφήνω σε σχόλια.
    transform = x * np.log1p(normalized) # Λύση 2 με log, παίρνω καλύτερα αποτελέσματα απο τη Λύση 1 αλλά και πάλι όχι ικανοποιητικά
    transform_img = np.uint8(np.clip(transform * 255 , 0, 255))  # Μετατροπή εικόνας ξανά σε τιμές 0-255 με uint8 
    return transform_img
restored = power_transformation(input_img, x)  # Αποθηκεύει στη restored την μετασχηματισμένη εικόνα
# Μετά από πολλά πειράματα κατέληξα στο γεγονός πως ο κώδικας που είναι ασχολίαστος καθώς και η τιμή 0.85 είναι η καταλληλότερη τιμή για την παράμετρο
# που έθεσα στην αρχή καθώς αυτά τείνουν πιο πολύ στους δεικτές της μη παραμορφοποιημένης εικόνας, ουσιαστικά η Λύση 2 κάνει s = c*log(1+r), οπότε
# πρέπει c<1 ώστε το s να μειωθεί και η εικόνα να γίνει πιο σκοτεινή εφόσον η distorted είναι υπερβολικά φωτεινή.
# === Οπτική αξιολόγηση ===
show_image("Distorted Image #4", input_img)
show_image("Restored Image", restored)

# === Ποσοτική αξιολόγηση μετά την αποκατάσταση (με χρήση της lena.jpg ως ground truth) === 
mse_score = np.mean((target - restored) ** 2)
psnr_score = psnr(target, restored)
ssim_score = ssim(target, restored)
print("After Restoration:")
print("MSE:", mse_score)
print("PSNR:", psnr_score)
print("SSIM:", ssim_score)
```

```python
## Αποκατάσταση: Distorted Image #5
# Η εικόνα έχει δεχθεί λειτουργία χαμηλής αντίθεσης καθώς είναι αρκετά σκοτεινή απο την κανονική, δηλαδή η απεικόνιση του ιστογράμματος της θα έδειχνε 
# τις τιμές συγκεντρωμένες στις μεσαίες τιμές του οριζόντιου άξονα του ιστογράμματος. Επομένως, μπορούμε να εφαρμόσουμε τεχνική ισοστάθμισης του
# ιστογράμματος για την αποκατάσταση της εικόνας.
input_img = distorted_images["Distorted Image #5"].copy()

# === Πριν την αποκατάσταση: Αξιολόγηση "distorted" ===
mse_distorted = np.mean((target - input_img) ** 2)
psnr_distorted = psnr(target, input_img)
ssim_distorted = ssim(target, input_img)
print("Before Restoration:")
print("MSE:", mse_distorted)
print("PSNR:", psnr_distorted)
print("SSIM:", ssim_distorted)

# === Εφαρμογή τεχνικών αποκατάστασης ===
def histogram_equalization(source_img):
    equalized_img = cv2.equalizeHist(input_img)   # Ισοστάθμιση ιστογράμματος με έτοιμη συνάρτηση απο την εκφώνηση, ουσιαστικά ανακατανέμει τις τιμές
                                                  # των pixel ομοιόμορφα σε όλο το εύρος των 0-255
    return equalized_img
restored = histogram_equalization(input_img)  # Αποθηκεύει τη αποκατεστημένη εικόνα στο restored για να τυπωθεί

# === Οπτική αξιολόγηση ===
show_image("Distorted Image #5", input_img)
show_image("Restored Image", restored)

# === Ποσοτική αξιολόγηση μετά την αποκατάσταση (με χρήση της lena.jpg ως ground truth) ===
# Ο δείκτης MSE, παρόλο που η εικόνα παρουσίασε σημαντική βελτίωση, αντί να μειωθεί φαίνεται να αυξήθηκε και αυτό οφείλεται στο ότι όταν έχουμε 
# ισοστάθμιση ιστογράμματος οι φωτεινότητες μεταβάλλονται απότομα για να καλύψει τις αντιθέσεις με αποτέλεσμα να μη διατηρεί ακριβώς τις αρχικές
# φωτεινότητες. Με την ισοστάθμιση οι σκούρες περιοχές γίνονται πιο σκούρες (βλέπε το ξύλο του καθρέπτη στην εικόνα) και οι φωτεινές περιοχές γίνονται
# πιο φωτεινές (βλέπε τον ώμο της Λένα) έτσι η διαφορά τετραγωνίζεται στο δείκτη MSE και αυξάνεται παρόλο που τελικά έχουμε βελτιωμένη εικόνα.
mse_score = np.mean((target - restored) ** 2)
psnr_score = psnr(target, restored)
ssim_score = ssim(target, restored)
print("After Restoration:")
print("MSE:", mse_score)
print("PSNR:", psnr_score)
print("SSIM:", ssim_score)
```
